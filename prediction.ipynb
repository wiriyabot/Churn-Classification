```python
import tensorflow as tf
import tensorflow.keras.models as load_model
import pickle
import pandas as pd
import numpy as np
```

Load the trained model, scaler pickle,onehet

```python
model = load_model('model.h5')
```

Load the encoder and scaler

```python
with open('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python
onehot_encoder_geo.categories_[0]
```

```python
label_encoder_gender.classes_
```

Example input data

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_encoder = onehot_encoder_geo.transform([[input_data['Geography']]]).toarray()
df_geo_encoder = pd.DataFrame(geo_encoder,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
df_geo_encoder
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

Encode categorical variables

```python
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

Concatination one hot encoded 

```python
input_df=pd.concat([input_df.drop("Geography",axis=1),df_geo_encoder],axis=1)
input_df
```

Scaling the input data

```python
input_scaled = scaler.transform(input_df)
input_scaled
```

Predict churn

```python
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print('The customer is not likely to churn.')
```
